```python
# 작업자 : 권은이

import pandas as pd
import json
import os

# --- 설정 ---
# ❗️ 여기에 변환하고 싶은 원본 CSV 파일 이름을 넣으세요.
CSV_FILE_NAME = 'et_reviews_최종.csv'
# --------------------

# JSONL 파일 이름은 CSV 파일 이름에 맞춰서 자동으로 설정됩니다.
JSONL_FILE_NAME = CSV_FILE_NAME.replace('.csv', '.jsonl')

# 파일 존재 여부 확인
if os.path.exists(CSV_FILE_NAME):
    print(f"✅ [준비 완료] 다음 파일을 변환합니다: '{CSV_FILE_NAME}'")
    print(f" -> JSONL 파일로 저장될 이름: '{JSONL_FILE_NAME}'")
else:
    print(f"🚨 [오류] '{CSV_FILE_NAME}' 파일을 찾을 수 없습니다!")
    print(" -> 파일 이름이나 경로를 다시 확인해주세요.")
```

```python
# 'df' 변수를 미리 초기화 (오류 방지)
df = None 

if os.path.exists(CSV_FILE_NAME):
    try:
        # encoding='utf-8-sig' : 한글 및 CSV의 BOM 문자 처리
        df = pd.read_csv(CSV_FILE_NAME, encoding='utf-8-sig')
        
        print(f"[성공] '{CSV_FILE_NAME}' 파일 로드 완료")
        print(f" -> 총 {len(df)} 개의 행(row)을 발견했습니다.")
        
        print("\n--- 데이터 샘플 (상위 5개) ---")
        display(df.head()) # 'display'는 Jupyter에서 표를 예쁘게 보여줌

        print("\n--- 데이터 기본 정보 (컬럼, 타입, Null 개수) ---")
        df.info()

    except Exception as e:
        print(f"[오류] 파일을 읽는 중 문제가 발생했습니다: {e}")
else:
    print("[오류] (셀 1)에서 파일이 확인되지 않았습니다. 셀 1을 먼저 실행해주세요.")
```

```python
if df is not None:
    print("--- 컬럼별 결측치(Null/NaN) 개수 확인 ---")
    
    null_counts = df.isnull().sum()
    
    if null_counts.sum() == 0:
        print("✅ [완벽] 모든 컬럼에 비어있는 데이터(Null)가 없습니다.")
    else:
        print("⚠️ [주의] 비어있는 데이터가 발견되었습니다:")
        # 비어있는 데이터가 1개 이상인 컬럼만 출력
        print(null_counts[null_counts > 0])
else:
    print("[오류] 'df'가 로드되지 않았습니다. (셀 2 실행 필요)")
```

```python
if df is not None:
    try:
        df.to_json(
            JSONL_FILE_NAME, 
            orient='records', 
            lines=True, 
            force_ascii=False,
            index=False # DataFrame의 인덱스(0, 1, 2...)는 저장하지 않음
        )
        
        print(f"✅ [성공] '{JSONL_FILE_NAME}' 파일 저장 완료!")
        print(f" -> 총 {len(df)} 개의 행이 JSONL로 변환되었습니다.")

    except Exception as e:
        print(f"[오류] JSONL 파일 저장 중 문제가 발생했습니다: {e}")
else:
    print("[오류] 'df'가 로드되지 않았습니다. (셀 2 실행 필요)")
```

```python
if os.path.exists(JSONL_FILE_NAME):
    try:
        print(f"--- (검증) 저장된 '{JSONL_FILE_NAME}' 파일 내용 (상위 5줄) ---")
        
        with open(JSONL_FILE_NAME, 'r', encoding='utf-8') as f:
            for i in range(5):
                line = f.readline()
                if not line: # 파일이 5줄보다 짧으면 중단
                    break
                print(line, end='') # line 자체에 \n이 있으므로 print는 줄바꿈 안 함
                
        if i == 0 and 'line' not in locals():
            print(" -> 파일이 비어있습니다. (변환된 데이터가 0건일 수 있습니다)")

    except Exception as e:
        print(f"[오류] JSONL 파일을 읽는 중 문제가 발생했습니다: {e}")
else:
    print(f"[오류] '{JSONL_FILE_NAME}' 파일이 생성되지 않았습니다. (셀 4 실행 확인)")
```
